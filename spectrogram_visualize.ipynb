```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

def plot_waveform(y, sr):
    plt.figure(figsize=(12, 4))
    librosa.display.waveshow(y, sr=sr)
    plt.title('Amplitude over Time')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

def plot_mel_spectrogram(y, sr):
    plt.figure(figsize=(12, 4))
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    S_dB = librosa.power_to_db(S, ref=np.max)
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', cmap='cividis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')
    plt.xlabel('Time (s)')
    plt.ylabel('Mel Frequency (Hz)')
    plt.show()

def plot_mfcc(y, sr):
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=23, lifter=1000)
    plt.figure(figsize=(12, 4))
    librosa.display.specshow(mfccs, sr=sr, x_axis='time')
    plt.colorbar()
    plt.title('MFCC')
    plt.xlabel('Time (s)')
    plt.ylabel('MFCC Coefficients')
    plt.show()

def main(audio_file):
    y, sr = librosa.load(audio_file, sr=None)
    y = y[:sr*5]  # Keep only the first 5 seconds
    plot_waveform(y, sr)
    plot_mel_spectrogram(y, sr)
    plot_mfcc(y, sr)

if __name__ == "__main__":
    audio_file = '/content/et_godt_spørsmål.wav'  # Replace with audio file path
    main(audio_file)
#https://matplotlib.org/stable/users/explain/colors/colormaps.html
#https://www.youtube.com/watch?v=g8Q452PEXwY
```
