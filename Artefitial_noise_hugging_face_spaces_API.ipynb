````python
#                                                   #
#   Create base sound + noise                       #
#   10% increments from 50% to 200% added noise     #
#                                                   #

import os
import shutil
from pydub import AudioSegment

# Paths to your directories
speech_directory = '/content/gdrive/My Drive/Master_2024_audio/artificially_noisy_speech/speech'
noise_directory = '/content/gdrive/My Drive/Master_2024_audio/artificially_noisy_speech/noise'
combined_directory = '/content/gdrive/My Drive/Master_2024_audio/artificially_noisy_speech/combined_noise_speech'

# Function to mix audio files
def mix_audio(original_path, noise_path, noise_level, output_path):
    # Load the original audio
    original_audio = AudioSegment.from_file(original_path)

    # Load the noise audio
    noise_audio = AudioSegment.from_file(noise_path)

    # If the noise audio is shorter than the original, loop it
    if len(noise_audio) < len(original_audio):
        noise_audio = noise_audio * (len(original_audio) // len(noise_audio) + 1)

    # Adjust the volume of the noise based on the noise level
    adjusted_noise = noise_audio - (100 - noise_level)

    # Overlay the adjusted noise onto the original audio
    combined = original_audio.overlay(adjusted_noise)

    # Export the mixed audio to a file
    combined.export(output_path, format="wav")

# Clear the combined_noise_speech directory before starting
if os.path.exists(combined_directory):
    shutil.rmtree(combined_directory)
os.makedirs(combined_directory, exist_ok=True)

# Iterate over audio files to transcribe
for audio_file in os.listdir(speech_directory):
    if audio_file.endswith('.wav'):
        # Full path to the original audio file
        audio_file_path = os.path.join(speech_directory, audio_file)

        # Iterate over noise files
        for noise_file in os.listdir(noise_directory):
            if noise_file.endswith('.wav'):
                # Create a directory for this combination of audio and noise type
                audio_file_name = os.path.splitext(audio_file)[0]
                noise_file_name = os.path.splitext(noise_file)[0]
                output_directory = os.path.join(combined_directory, f"{audio_file_name}_{noise_file_name}")
                os.makedirs(output_directory, exist_ok=True)

                # Full path to the noise file
                noise_file_path = os.path.join(noise_directory, noise_file)

                # Initialize a counter for numbering the files
                file_counter = 1

                # Mix and save audio with noise at levels from 50% to 200% in 10% increments
                # Loop for 40% to 80% with 10% increments
                for level in range(40, 80, 10):
                    output_file_name = f"{file_counter:03d}_{audio_file_name}_{level}%_{noise_file_name}_noise.wav"
                    output_file_path = os.path.join(output_directory, output_file_name)
                    # Assume mix_audio is a function that mixes audio and noise at the given level and saves to output_file_path
                    mix_audio(audio_file_path, noise_file_path, level, output_file_path)
                    print(f"Generated {output_file_path}")
                    file_counter += 1

                # Loop for 80% to 110% with 4% increments
                for level in range(80, 120, 4):  # Adjusted the stop value to 115 to include 110%
                    output_file_name = f"{file_counter:03d}_{audio_file_name}_{level}%_{noise_file_name}_noise.wav"
                    output_file_path = os.path.join(output_directory, output_file_name)
                    mix_audio(audio_file_path, noise_file_path, level, output_file_path)
                    print(f"Generated {output_file_path}")
                    file_counter += 1

                # Loop for 110% to 140% with 10% increments
                for level in range(120, 140, 10):
                    output_file_name = f"{file_counter:03d}_{audio_file_name}_{level}%_{noise_file_name}_noise.wav"
                    output_file_path = os.path.join(output_directory, output_file_name)
                    mix_audio(audio_file_path, noise_file_path, level, output_file_path)
                    print(f"Generated {output_file_path}")
                    file_counter += 1
````

# Transcribe using Whisper Large-V3
Audio from combined_noise_speech is sent to the whisper Large huggingface space provided by OpenAi. The results are stored in results/. Each model has its own folder and each speech creates a new folder if there is no folder existing.

│   │   ├── 015_speech_A_130%_Noise_A_transcribed.txt
│   │   └── 016_speech_A_140%_Noise_A_transcribed.txt

````python
import os
from gradio_client import Client, file
from concurrent.futures import ThreadPoolExecutor, TimeoutError

# Configuration for different Gradio client models
models = [
    {"id": "kjetMol/openai-whisper-large-v3", "sub_dir": "openai-whisper-large-v3"},
    {"id": "kjetMol/NbAiLabBeta-nb-whisper-large", "sub_dir": "NbAiLab-nb-whisper-large"},
    {"id": "kjetMol/NbAiLab-nb-whisper-large-verbatim", "sub_dir": "NbAiLab-nb-whisper-large-verbatim"},
]

base_dir = "/content/gdrive/My Drive/Master_2024_audio/artificially_noisy_speech"
noise_dir = os.path.join(base_dir, "noise")
audio_dir = os.path.join(base_dir, "combined_noise_speech")

def setup_directories(base_transcribe_dir):
    os.makedirs(base_transcribe_dir, exist_ok=True)
    for noise_file in os.listdir(noise_dir):
        noise_type = os.path.splitext(noise_file)[0]
        os.makedirs(os.path.join(base_transcribe_dir, noise_type), exist_ok=True)

def transcribe_audio(client, file_path, timeout_seconds=45):
    def task():
        print(f"Transcribing {file_path}")
        return client.predict(file(file_path), api_name="/predict")

    with ThreadPoolExecutor(max_workers=3) as executor:  # Using a separate executor for each transcription task
        future = executor.submit(task)
        try:
            result = future.result(timeout=timeout_seconds)
            print(f"Result for {file_path}: {result}")
        except TimeoutError:
            print(f"Transcription timed out for {file_path}")
            return None

    if isinstance(result, str):
        return result
    else:
        return None

def get_transcription_folder(file_name, base_transcribe_dir):
    for noise_file in os.listdir(noise_dir):
        if os.path.splitext(noise_file)[0] in file_name:
            return os.path.join(base_transcribe_dir, os.path.splitext(noise_file)[0])
    return base_transcribe_dir

def process_audio_files(model_config):
    hf_token = os.getenv('HF_TOKEN')
    client = Client(model_config["id"], hf_token=hf_token)

    transcribe_dir_base = os.path.join(base_dir, "transcribed_audio", model_config["sub_dir"])
    setup_directories(transcribe_dir_base)

    successfully_transcribed = set()
    for dirpath, _, filenames in os.walk(audio_dir):
        for filename in filenames:
            if filename.endswith('.wav'):
                file_path = os.path.join(dirpath, filename)
                transcription_result = transcribe_audio(client, file_path)

                if transcription_result:
                    transcription_folder = get_transcription_folder(filename, transcribe_dir_base)
                    text_file_path = os.path.join(transcription_folder, f"{os.path.splitext(filename)[0]}_transcribed.txt")

                    with open(text_file_path, 'w') as text_file:
                        text_file.write(transcription_result)
                    successfully_transcribed.add(text_file_path)
                    print(f"Transcription for {filename} saved to {text_file_path}")

# Running all models concurrently
with ThreadPoolExecutor(max_workers=len(models)) as executor:
    futures = [executor.submit(process_audio_files, model) for model in models]
    for future in futures:
        future.result()  # Wait for each future to complete

````

````python
import os

# Configuration for different Gradio client models
models = [
    {"id": "kjetMol/openai-whisper-large-v3", "sub_dir": "openai-whisper-large-v3"},
    {"id": "kjetMol/NbAiLabBeta-nb-whisper-large", "sub_dir": "NbAiLab-nb-whisper-large"},
    {"id": "kjetMol/NbAiLab-nb-whisper-large-verbatim", "sub_dir": "NbAiLab-nb-whisper-large-verbatim"},
]

base_dir = "/content/gdrive/My Drive/Master_2024_audio/artificially_noisy_speech"
audio_dir = os.path.join(base_dir, "combined_noise_speech")

def get_transcription_folder(file_path, base_transcribe_dir):
    # Extract parts from the filename using os.path.split and slicing
    parts = file_path.split('/')
    # Typical file path:
    # /.../fortsatte_sende_filer_angry_male_crowd_ambience/001_fortsatte_sende_filer_40%_angry_male_crowd_ambience_noise.wav
    file_name_parts = parts[-1].split('_')
    identifier = file_name_parts[1]  # Identifier is after the first '_'
    noise_type_parts = parts[-2].split('_')
    noise_type = '_'.join(noise_type_parts[-4:])  # Join the last three parts to form the noise type

    # Create a folder path with both noise type and identifier
    target_folder = os.path.join(base_transcribe_dir, noise_type, identifier)
    os.makedirs(target_folder, exist_ok=True)  # Ensure the directory exists
    return target_folder

def simulate_transcription_process():
    for model_config in models:
        transcribe_dir_base = os.path.join(base_dir, "transcribed_audio", model_config["sub_dir"])

        print(f"Simulating for model: {model_config['id']}")
        for dirpath, _, filenames in os.walk(audio_dir):
            for filename in filenames:
                if filename.endswith('.wav'):
                    file_path = os.path.join(dirpath, filename)
                    print(f"Would transcribe {file_path}")

                    transcription_folder = get_transcription_folder(file_path, transcribe_dir_base)
                    text_file_path = os.path.join(transcription_folder, f"{os.path.splitext(filename)[0]}_transcribed.txt")

                    print(f"Would save transcription for {filename} to {text_file_path}")

simulate_transcription_process()
````
