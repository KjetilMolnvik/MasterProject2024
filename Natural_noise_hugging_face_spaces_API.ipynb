# Convert m4a to wav

````python
import os
import shutil
from pydub import AudioSegment

def clear_directory(directory_path):
    """Remove all files in the specified directory."""
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print(f'Failed to delete {file_path}. Reason: {e}')

def convert_and_split_audio(source_dir, target_dir, segment_length_ms=20000):
    """
    Clear the target directory, convert all audio files in the source directory to WAV format,
    split them into segments, and save them in the target directory.

    Args:
    - source_dir: Path to the source directory containing audio files.
    - target_dir: Path to the target directory where segmented WAV files will be saved.
    - segment_length_ms: Length of each segment in milliseconds. Default is 20000ms (20 seconds).
    """
    # Clear the target directory first
    clear_directory(target_dir)

    # Ensure the target directory exists
    os.makedirs(target_dir, exist_ok=True)

    # Iterate through all files in the source directory
    for filename in os.listdir(source_dir):
        if not filename.endswith(('.m4a', '.mp3', '.flac', '.ogg', '.wav')):  # Skip non-audio files
            continue

        # Determine the format of the audio file based on its extension
        file_format = filename.split('.')[-1]
        basename = os.path.splitext(filename)[0]

        # Full path to the original file
        original_file = os.path.join(source_dir, filename)

        # Load the original audio file
        sound = AudioSegment.from_file(original_file, format=file_format)

        # Split and export the segments
        for i in range(0, len(sound), segment_length_ms):
            segment = sound[i:i + segment_length_ms]
            segment_file_path = os.path.join(target_dir, f"{basename}_part_{i // segment_length_ms}.wav")
            segment.export(segment_file_path, format="wav")
            print(f"Exported {segment_file_path}")

source_dir = '/content/gdrive/My Drive/Master_2024_audio/naturally_noisy_speech/noisy_speech'
target_dir = '/content/gdrive/My Drive/Master_2024_audio/naturally_noisy_speech/noisy_speech_converted'
convert_and_split_audio(source_dir, target_dir)
````

# Transcribe using Whisper Large-V3

````python
import os
from gradio_client import Client, file
from concurrent.futures import ThreadPoolExecutor, TimeoutError

# Configuration for different Gradio client models
models = [
    {"id": "kjetMol/openai-whisper-large-v3", "sub_dir": "openai-whisper-large-v3"},
    {"id": "kjetMol/NbAiLabBeta-nb-whisper-large", "sub_dir": "NbAiLab-nb-whisper-large"},
    {"id": "kjetMol/NbAiLab-nb-whisper-large-verbatim", "sub_dir": "NbAiLab-nb-whisper-large-verbatim"},
]

# Updated base directory for naturally noisy speech
base_dir = "/content/gdrive/My Drive/Master_2024_audio/naturally_noisy_speech"
audio_dir = os.path.join(base_dir, "noisy_speech_converted")  # Directory containing the audio files

def setup_directories(base_transcribe_dir):
    os.makedirs(base_transcribe_dir, exist_ok=True)

def transcribe_audio(client, file_path, timeout_seconds=100):
    def task():
        print(f"Transcribing {file_path}")
        return client.predict(file(file_path), api_name="/predict")

    with ThreadPoolExecutor(max_workers=1) as executor:
        future = executor.submit(task)
        try:
            result = future.result(timeout=timeout_seconds)
            print(f"Result for {file_path}: {result}")
        except TimeoutError:
            print(f"Transcription timed out for {file_path}")
            return None

    if isinstance(result, str):
        return result
    else:
        return None

def process_audio_files(model_config):
    hf_token = os.getenv('HF_TOKEN')
    client = Client(model_config["id"], hf_token=hf_token, upload_files=True)

    # Directory where transcription results will be stored
    transcribe_dir_base = os.path.join(base_dir, "transcription", model_config["sub_dir"])
    setup_directories(transcribe_dir_base)

    for filename in os.listdir(audio_dir):
        if filename.endswith('.wav'):
            file_path = os.path.join(audio_dir, filename)
            transcription_result = transcribe_audio(client, file_path)

            if transcription_result:
                text_file_path = os.path.join(transcribe_dir_base, f"{os.path.splitext(filename)[0]}_transcribed.txt")

                with open(text_file_path, 'w') as text_file:
                    text_file.write(transcription_result)
                print(f"Transcription for {filename} saved to {text_file_path}")

# Running all models concurrently
with ThreadPoolExecutor(max_workers=len(models)) as executor:
    futures = [executor.submit(process_audio_files, model) for model in models]
    for future in futures:
        future.result()  # Wait for each future to complete
````
